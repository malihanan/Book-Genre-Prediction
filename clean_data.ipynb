```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
import nltk
import csv
import pandas as pd
import json
import re
from nltk.corpus import stopwords
from nltk import WordNetLemmatizer
import math
```

```python
stop_words = stopwords.words('english')
wnl = WordNetLemmatizer()
def clean_data(p):
    dataset = nltk.sent_tokenize(p)
    words = [word for data in dataset for word in nltk.word_tokenize(data)]
    words = [wnl.lemmatize(item[0].lower()) for item in nltk.pos_tag(words) if item[1] != 'NNP' and item[0].lower() not in stop_words]
    p = ' '.join(words)

    t = ""
    for x in nltk.sent_tokenize(p):
        x = x.lower()
        x = re.sub(r'\W', ' ', x)
        x = re.sub(r'\s+', ' ', x)
        x = re.sub(r'^\s+', '', x)
        x = re.sub(r'\s$', '', x)
        t += ". " + x
    return t[2:]
```

```python
def getList(x):
    j = json.loads(x)
    l = []
    for i in j:
        x = clean_data(j[i])
        l.append(x)
    return l
```

```python
filename = "booksummaries1.txt"
        
dataset=[]
with open(filename, 'r', encoding="utf8", errors='ignore') as f:
    reader=csv.reader(f,dialect='excel-tab')
    for row in reader:
        dataset.append([row[0],row[2],row[3],row[4],row[5],row[6]])
df=pd.DataFrame(dataset,columns=['ID','Name','Author','Date','Genres','Summary'])

df=df.drop(df[df['Genres']==''].index)
df['Genres'] = df['Genres'].apply(lambda x: getList(x))
df['Summary'] = df['Summary'].apply(lambda x: clean_data(x))
```
